```python
import holoviews as hv
hv.extension('bokeh')
```

```python
hv_ds = hv.Dataset(ds_hci)
hv_ds
```

```python
from holoviews import opts
curves = hv_ds.to(hv.Curve, 'start_time').overlay().redim(Year=dict(range=(1970, 2030)))
curves.opts(opts.Curve(width=800, height=550, framewise=True))
```

```python

layout = hv.Curve((ds_hci.start_time.data, ds_hci.hci.isel(lead_time=0).data))
layout.opts(opts.Curve(width=800, height=250, framewise=True))
```

```python
def loop_lead(lead):
    return hv.Curve((ds_hci.start_time.data, ds_hci.hci.sel(lead_time=lead).data))
dyn_hci = hv.DynamicMap(loop_lead, kdims=[hv.Dimension('lead_time', values=ds_hci.lead_time.data)])
dyn_hci.opts(opts.Curve(width=500, height=250, framewise=True))
dyn_hci.redim.range(lead_time=(0.5,11.5), y=(-0.1,1.1))
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(1)
ax1=fig.add_axes([0,0,3,1])
lead_time = [0.5,6.5,11.5]
colors = ['#E88D67','#059212','#3572EF']

for nlead,leadtime in enumerate(lead_time):
    print(leadtime)
    time = ds_hci.sel(start_time=slice('2013-01','2017-12')).start_time.data
    da_hci = ds_hci.sel(lead_time=leadtime).hci
     
    if leadtime == 0.5:
        da_hci = da_hci.sel(start_time=slice('2013-01','2017-12'))
    elif leadtime == 6.5:
        da_hci = da_hci.sel(start_time=slice('2012-07','2017-06'))
    elif leadtime == 11.5:
        da_hci = da_hci.sel(start_time=slice('2012-02','2017-01'))
    da_hci['start_time'] = time

    (
    da_hci
    .plot(
        color=colors[nlead],
        linestyle='solid',
        linewidth=3,
        ax=ax1,
        label=f'lead time {leadtime}',
        alpha=0.9,
    )
    )

    mean = (
    da_hci
    .mean('start_time')
    )
    ax1.axhline(
        y=mean,
        linestyle='dashed',
        linewidth=2,
        alpha=0.9,
        color=colors[nlead],
    )


    #### setting the plotting format
    ax1.set_ylabel('HCI',{'size':'25'},color='k')
    ax1.tick_params(axis='y',labelsize=20,labelcolor='k')
    ax1.tick_params(axis='x',labelsize=20,labelcolor='k',rotation=20)
    ax1.set_xlabel('Year',{'size':'25'},labelpad=20)
    ax1.set_ylim([-0.1,1])
    ax1.set_title("Habitat Compression Index",{'size':'24'},pad=24)
    ax1.legend(loc='lower left',fontsize=14,frameon=False)
```
